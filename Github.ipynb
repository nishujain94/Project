```python
b = df.groupby(['spec'])['spec'].count()
c = pd.DataFrame(b)
b
c
```

```python
d = b.nlargest(10)
d
e = pd.DataFrame(d)
e['count'] = e['spec']
del e['spec']
e
```

```python
e['spec'] = e.index
e
```

```python
x = e['spec']
y = e['count']
plt.barh(x, y, color = 'g')
```
